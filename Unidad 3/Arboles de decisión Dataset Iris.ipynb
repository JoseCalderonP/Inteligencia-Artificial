```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn import tree

iris = load_iris()
iris
```

```python
clf = tree.DecisionTreeClassifier()
# Primer parametro son los valores de las medidas de las flores.
# Segundo parametro es la variable que se quiere predecir.
clf = clf.fit(iris.data, iris.target)

clf.score(iris.data, iris.target)
#Genera la salida en formato .dot
#Para poder visualizar el archivo dot con la libreria graphviz
tree.plot_tree(clf)
```

```python
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph.render('iris', format='png')
```
